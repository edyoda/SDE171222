# append() in python

- inserts a single element into an existing list.
- the element will be added to the end of the old list rather than being returned to a new list.
- the length of the list increases by 1.

```python
l = [10, 20, "as it is", 3.0]
l.append(6)
print(l)
```

```python
l1 = [10, 20, "as it is", 3.0]
l2 = [1, 2, 3]
l1.append(l2)
print(l1)
print(l1[4])
print(l1[4][1])
```

```python
l1 = [10, 20, "apple", 3.0]
l2 = [1, 2, 3]
l1.append(l2)
print(l1)
print(l1[4])
print(l1[4][1])
print(l1[2][3])
```

# extend in python

- iterates over its arguments and adding each element to the list and extending the list. 
- The length of the list increases by the number of elements in its arguments.

```python
l1 = [10, 20, "apple", 3.0]
l2 = [1, 2, 3]
l1.extend(l2)
print(l1)
```

# sort() method:

- sort method sorts the list ascending by default.
- key: (optional) a function to specify the sorting criteria

```python
cars = ['Honda', 'Ford', 'BMW', 'Volvo', 'Kia']
cars.sort()
print(cars)
```

```python
cars = ['Honda', 'Ford', 'BMW', 'Volvo', 'Kia']
cars.sort(reverse=True)
print(cars)
```

```python
cars = ['Honda', 'Ford', 'BMW', 'Volvo', 'Kia']
cars.sort(reverse=False)
print(cars)
```

```python
def count_char(word):
    return len(word)

cars = ['Honda', 'Ford', 'BMW', 'Volvo', 'Kia']
cars.sort(key=count_char)
print(cars)
```

```python
def count_char(word):
    return len(word)

cars = ['Honda', 'Ford', 'BMW', 'Volvo', 'Kia']
cars.sort(reverse=True, key=count_char)
print(cars)
```

Example of list of dictionaries:

```python
def sort_on_year(list_of_cars):
    return list_of_cars['year']

cars = [
    {'car': 'Ford', 'year': 2000},
    {'car': 'Kia', 'year': 2015},
    {'car': 'BMW', 'year': 1998},
    {'car': 'VW', 'year': 2011}
]

cars.sort(key=sort_on_year)
print(cars)
```

# Count the number of vowels in list

```python
l = ['a', 'w', 'e', 'r']
count_a = l.count('a')
count_e = l.count('e')  
count_i = l.count('i')
count_o = l.count('o')
count_u = l.count('u') 
print(count_a + count_e + count_i + count_o + count_u)
```

```python
l = ['a', 'w', 'e', 'r']
vowels = ['a', 'e', 'i', 'o', 'u']

count = 0
for vowel in vowels:
    if vowel in l:
        count += 1
        
print(count)
```

# Count number of strings where string length is 2 or more and the first & last character are same from a given list of strings

```python
list_of_strings = ['abc', 'xyz', 'abba', 'af','gj', '1341', '576', '66']
```

```python
count = 0

for word in list_of_strings:
    if len(word) >= 2 and word[0] == word[-1]:
        count += 1
print(count)
```

creation function for same task:

```python
def count_specific_strings(list_of_strings):
    count = 0

    for word in list_of_strings:
        if len(word) >= 2 and word[0] == word[-1]:
            count += 1
    return count
```

```python
x = count_specific_strings(['abc', 'xyz', 'abba', 'af','gj', '1341', '576', '66'])
print(x)
```

```python
x = count_specific_strings(['12'])
print(x)
```
