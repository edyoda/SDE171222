# List Comprehension

- consists of square brackets containing the expression, which is executed for each element along with the for loop 

Syntax:
    
[expression for item in iterable]


```python
s = "human"

ol = []
for i in s:
    ol.append(i)
print(ol)
```

```python
ol2 = [i for i in s]
print(ol2)
```

```python
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]

# if 'a' is present then append the fruit name to a new list

new_list = []
for i in fruits:
    if "a" in i:
        new_list.append(i)
        
print(new_list)
```

```python
new_list2 = [i for i in fruits if "a" in i]
print(new_list2)
```

```python
# add all fruits names which aren't apple in output list

ol = [i for i in fruits if i != "apple"]
print(ol)
```

```python
numbers = [2, 3, 4, 5]

# create a list contatining even numbers 
new_list = []
for num in numbers:
    if num % 2 == 0:
        new_list.append(num)
print(new_list)
        
new_list2 = [num for num in numbers if num % 2 == 0]
print(new_list2)
```

```python
[x + 1 for x in range(5) if x % 2 == 2]
```

```python
[x + 1 for x in range(5) if x % 2 == 1]
```

```python
l = [["apple", "banana"]]

for i in l:                    
    print(i)
    for j in i:
        print("j", j)
```

# Nested list comprehension

- Nested list comprehension are list comprehension within another list comprehension which is quite similar to nested for loops. 

```python
li = [3, 2, 4, 1]
[element * 2 for element in [item + 1 for item in li]]
```

```python
matrix = []

for i in range(3):
    #append an empty list inside the list
    matrix.append([])
    
    for j in range(5):
        matrix[i].append(j)
        
    print("row:", matrix[i])
        
print(matrix)
```

# Time Analysis in List Comprehension and loop:


```python
import time

def for_loop(n):
    result = []
    for i in range(n):
        result.append(i**2)
    return result

def list_comprehension(n):
    return [i ** 2 for i in range(n)]


begin = time.time()
for_loop(10**6)
end = time.time()
print("time taken for FOR LOOP execution:", round(end-begin, 2))

begin = time.time()
list_comprehension(10**6)
end = time.time()
print("time taken for list comprehension execution:", round(end-begin, 2))
```
