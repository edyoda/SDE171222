# Find peak element which is not smaller than its neigbors

ia = [5, 10, 20, 15]

output = 2 (index of 20)

```python
def findpeakElement(nums):
    left=0
    right=len(nums)-1
    while left < right:
        mid = (left+right) // 2
        if nums[mid]> nums[mid+1]:
            right = mid
        else:
            left =mid +1
    return left
```

```python
findpeakElement([5, 10, 20, 15, 98])
```

```python
def find_peak(arr):
    
    if len(arr)== 1:
        return 0
    
    if(arr[0] >= arr[1]):
        return 0
    
    if (arr[-1] >= arr[-2]):
        return len(arr)-1
    
    for i in range(1, len(arr)-1):
        if (arr[i] >= arr[i-1] and arr[i] >= arr[i+1]):
            return i
    
```

```python
find_peak([5, 10, 20, 15, 98])
```

```python
def find_all_peak_indices(arr):
    
    peak_indices = []
    
    if len(arr)== 1:
        peak_indices.append(0)
    
    if(arr[0] >= arr[1]):
        peak_indices.append(0)
    
    if (arr[-1] >= arr[-2]):
        peak_indices.append(len(arr)-1)
    
    for i in range(1, len(arr)-1):
        if (arr[i] >= arr[i-1] and arr[i] >= arr[i+1]):
            peak_indices.append(i)
            
    return peak_indices
    
```

```python
find_all_peak_indices([5, 10, 20, 15, 98])
```
