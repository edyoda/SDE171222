# Methods/ functions of Python string

upper():

- converts all lowercase characters in the string into uppercase characters and returns it

```python
msg = "python is fun"
print(msg.upper())
```

```python
msg = "python is fun"
x = msg.upper()
print(x)
```

```python
msg = "pyTHon is fun"
x = msg.upper()
print(x)
```

lower()

- converts all uppercase characters in the string into lowercase characters and returns it
- find() method returns an integer value

```python
msg = "PYTHON IS FUN"
x = msg.lower()
print(x)
```

```python
msg = "PYTHON IS fun"
x = msg.lower()
print(x)
```

find():


this method returns the index of the first occurrence of the substring(if found). If not found, it returns -1.

```python
msg = "Python is an interprted programming language."
x = msg.find('program')
print(x)
```

```python
msg = "My name is"
x = msg.find("nam")
print(x)
```

```python
msg = "My name is"
x = msg.find("nam ")
print(x)
```

```python
quote = "Take small steps at once to acheive your greater goals"
x = quote.find('small steps', 10)
print(x)
```

```python
quote = "Take small steps at once to acheive your greater goals"
x = quote.find('small steps', 3)
print(x)
```

```python
x = quote.find('small steps', 3, 46)
print(x)
```

replace():


- it replaces each matching occurrence of a substring with another string
- return value: returns a copy of the string(new string) where the old substring is replaced with the new string. The original strings remains unchanged.
- if the old substring is not found, it returns a copy of the originaal string.

```python
text = "bat ball"
# replace 'ba' with 'ro'
replaced_text = text.replace('ba', 'ro')
print(replaced_text)
```

```python
text = "bat ball"
# replace 'ba' with 'ro'
replaced_text = text.replace('ba', 'ro#')
print(replaced_text)
```

```python
text = "bat ball"
print(id(text))
# replace 'ba' with 'ro'
text = text.replace('ba', 'ro#')
print(text)
print(id(text))
```

replace() arguments:

- old: the old substring we want to replace
- new: new substring which will replace the old substring
-count(optional): the no. of times u want to replace the old substring with the new string

```python
text = "bat ball"
replaced_text = text.replace('ba', 'ro#', 4)
print(replaced_text)
```

```python
text = "bat ball"
replaced_text = text.replace('bat ball', 'ro#')
print(replaced_text)
```

```python
txt = "babababa"

x = txt.replace("ba", "ro#" ,3)

print(x)
```

index():


- it returns the index of a substring inside the string(if found). If the substring is not found, it raises an exception.

```python
text = "I love Python"
res = text.index('th')
print(res)
```

```python
text = "I love Python"
res = text.index('z')
print(res)
```

index() parameters:

- sub: substring to be searched in the string str
- start and end(optional): substring is searched within str[start:end]

```python
sentence = "I love Python prograaming"
# len(sentence)
print(sentence.index('n pro', 10, 18))
```

```python
sentence = "I love Python prograaming"
# len(sentence)
print(sentence.index('n pro', 10, -1))
```

split() 

- splits a string into a list where each word is a list item

```python
txt = "welcome to python"
x = txt.split()
print(x)
```

syntax:

string.split(separator, maxsplit)

- separator: optional. Specifies the separator to use when splitting the string. By default, any whitespace is a separator
- maxsplit: optional. Specifies how many splits to do. By default, it does it for all occurrences.

```python
txt = "welcome to python, my name is Shikha, I'm a student"
x = txt.split(",")
print(x)
```

```python
txt = "apple@is@a@fruit"
x = txt.split("@")
print(x)
```

```python
txt = "apple#banana#grapes#cherry"
x = txt.split("#", 1)
print(x)
```
