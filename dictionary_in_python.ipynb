Q. What are the different ways of creating a dictionary?

```python
# Create an empty dictionary

d1 = {}
key = 'a'
value = 1
d1[key] = value
print(d1)

# Create dictionary using dict() method

d1 = dict({'a': 1, 'b': 2}) 
print(d1)

# Create dictionary with each item as pair

d1 = dict([('a', 1), ('b', 2)])
print(d1)

# The most popular way of defining dictionary
d1 = {'a': 1, 'b': 2}
print(d1)
```

Q. What is a nested dictionary?

```python
d1 = {1: {'roll_no': 101, 'name': 'ABC'},
     2: {'roll_no': 102, 'name': 'XYZ'}}
print(d1)
```

```python
d1[1]
```

```python
for key in d1:
    print(key, d1[key])
```

```python
d1 = {1: {'roll_no': 101, 'name': 'ABC'},
     2: {'roll_no': 102, 'name': 'XYZ'}}

for item in d1:                 #1                                                     #2 
    print(item, d1[item])       #1 #d1[1]                                              #d1[2]
    nested_dict = d1[item]      #nd = d1[1]
    for k in nested_dict:      #{'roll_no': 101, 'name': 'ABC'}
        print(k, nested_dict[k])  #nested_dict['roll_no']  #'name', nested_dict['name']
```

# Python program to find the sum of all items in a dictionary 

```python
d = {'x': 100, 'y': 20, 'z': 20}
output = 140
```

```python
d = {1: 100, 2: 200}
sum(d)
```

```python
sum(d.values())
```

```python
def sum_of_dict_values(d):
    sum_  = 0
    for i in d.values():
        #     sum_ = sum_ + i
        sum_ += i                 #100 #100+20 #100+20+20

    return sum_
```

```python
d = {'x': 100, 'y': 20, 'z': 20}
# output = 140
sum_of_dict_values(d)
```
