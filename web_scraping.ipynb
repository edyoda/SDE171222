# Two-sum problems

```python
inp=[10,15,2,7,39] #o/p:[0,2]
op=[]
target=12


for each in inp:
        element=target-each   #12-10=2
        if element in inp:
            op.append(inp.index(each))
            op.append(inp.index(element))
print(list(set(op))) 
```
