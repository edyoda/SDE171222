- mutable datatype
- they are unordered
- can have different datatypes

# Add and update set items

```python
# sets are unordered
s = {10, 20, 30}
print(s[0])
```

```python
numbers = {21, 45, 546, 78}
print("initial set: ", numbers)
print(id(numbers))

# append fn. isn't valid in sets bcoz sets are unordered, they don't have any starting or ending,
# but append fn. always appends the element at end
numbers.add(100)
print("updated set: ", numbers)
print(id(numbers))
```

```python
# append fn. isn't valid in sets bcoz sets are unordered, they don't have any starting or ending,
# but append fn. always appends the element at end
numbers.append(5)
```

```python
numbers = {21, 45, 546, 78}
numbers2 =  [100, 600, 45, 36, 498]

numbers.update(numbers2)
print(numbers)
```

# Remove an element from a set

```python
numbers = {21, 45, 546, 78}
print("inital set: ", numbers)
numbers.discard(45)
print("set after deleting: ", numbers)
```

# Python set operations

```python
A = {1, 3, 5}
B = {0 , 2, 4}
print("union of A and B: ", A.union(B))
print("union of A and B: ", A | B)
```

```python
A = {1, 3, 5, 4}
B = {0 , 2, 4}
print("intersection of A and B: ", A.intersection(B))
print("intersection of A and B: ", A & B)
```

```python
A = {1, 3, 5, 4}
B = {0 , 2, 4}
print("difference of A and B: ", A - B)
print("difference of B and A: ", B - A)
```
