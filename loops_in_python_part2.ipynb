# SLICING

```python
s = "This city"
print(s[2:7])
print(s[:5])
print(s[3:0:-1])
print(s[0:5:2])
print(s[3:0:-2])
```

```python
numbers = [1, 2, 3, 4, 5]
result = []
for num in numbers:
    if num % 2 == 0:
        result.append(num * 2)
    else:
        result.append(num)
```

```python
s = "This city"
for i in s:
    print(i, end="")
```

```python
s = "This city"

for i in s:
    if i in ['a', 'e', 'i', 'o', 'u']:
        print("z")
    elif i== "i" or i == "e":
        print("y")
    else:
        print(i)
```
