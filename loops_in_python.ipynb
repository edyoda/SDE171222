# range(start, stop, step)

there're 3 ways of writing range. All these 3 patterns are shown below:

```python
for i in range(4):
    print(i)
    
for i in range(10, 14):
    print(i)
    
for i in range(10, 14, 2):
    print(i)
```

```python
numbers = [10, 42, 56, 78]
for i in range(len(numbers)):
    print(i, numbers[i])
    
# OR-----------------------------

for i in numbers:
    print(i)
    
# OR-------------------------------

for index, num in enumerate(numbers):
    print(f'The index is: {index} and the number is: {num}')
```

```python
numbers = [2, 3, 4, 5, 6]

# range(start, stop, step)  
for i in range(len(numbers)):                #for i in range(5)   #0 to 4 #i=0 #i=1
    print("value of i", i)                   #i=0                              #i=1
    print("element at index i: ", numbers[i])   #numbers[0]                    #numbers[i]
```

```python
for i in range(10, 20):              #10 to 19
    print("value of i", i)
```

```python
for i in range(10, 20, 2):              #10 to 19
    print("value of i", i)
```

```python
for i in range(20, 10, -1):              #range from start,stop-1
    print("value of i", i)
```

# while loop

Syntax: 

while <condition>:
    <code_block>
        
NOTE: make sure the while loop doesn't runs into infinite loop

```python
counter = 0
while counter < 10:                              #1<10     #2<10
    print("counter", counter)     #0             #1        #2
    counter = counter + 1         #c=0+1=1       #c=1+1    #c=2+1
```

```python
counter = 1

while counter < 11:
    print(counter * 2)
    counter = counter + 1
```

# find factorial of a number

```python
def factorial_of_a_number(number):
    factorial = 1
    
    if number < 0:
        return "Factorial only valid for positive numbers"
    
    elif number == 0:
        return "1"
    
    else:
        for i in range(1, number+1):
            factorial = factorial * i
        
    return factorial
```

```python
x = factorial_of_a_number(5)
print(x)
```
