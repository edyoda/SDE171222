# Python Function declaration:


- def: keyword used to declare a function
- function_name: any name given to the function
- argumnets: values passed to the function
- return(optional): returns a value from a function

```python
# declaration of function
def greet():
    print("Hello World")
```

```python
#call the function
greet()
```

```python
def greet_with_name(name):
    print("Hello" + " " +  name)
```

```python
def greet_with_name_and_city(name, city):
    print("Welcome to city: ", city)
    print("Hello" + " " + name)
```

```python
greet_with_name_and_city("Shikha", "Bangalore", "xyz")
```

# Function Overloading:

- functions having the same name but different number of arguments

```python
def addition(n1, n2):
    return n1 + n2
```

```python
addition(8, 2)
```

```python
def addition(n1, n2, n3):
    return n1 + n2 + n3
```

```python
addition(8, 2, 4)
```

```python
addition(9, 18, 27)
```

```python
addition(8, 5)
```

# return 

```python
def alter_no(n):
    return (n ** 2) * 3
```

```python
x = alter_no(3)
print(x)
```

```python
def alter_no_v2(n):
    return (n ** 2) * 3, n
```

```python
x, y = alter_no_v2(3)
print(x, y)
```

# Functions with variable number of arguments:

```python
def add(*args):
    sum=0
    for num in args:
        sum += num
    return sum
```

```python
add(1, 2)
```

```python
add(1, 2, 3, 4)
```

```python
add(1, 2, 4, 5, 6, 7, 8)
```
