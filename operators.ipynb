# Assignment Operators

- assignment operators are used to assign values to variables.

```python
a = 10
b = 5
```

```python
a += b           # a = a + b
print(a)
```

```python
a -= b          # a = a - b
print(a)          # a=15, b = 5
```

```python
a *= b         # a = a * b
print(a)          #a = 75, b=5
```

```python
a /= b         # a = a / b
print(a)         # a = 15, b = 5
```

```python
a **= b      #a = a ** b
print(a)       #a= 759375, b = 5   
```

```python
a %= b       # a = a % b
print(a)
```

# Identity Operators

- are used to compare the variables, not if they are equal, but if they are actually sharing the same memory location.

```python
x = 10             #x == 10
y = 10
print(x is y)
```

```python
x3 = [1, 2, 3]
y3 = [1, 2, 3]

print(x3 is y3)
```

```python
x3 = [1, 2, 3]
y3 = [1, 2, 3, 4]

print(x3 is y3)
```

```python
x3 = [1, 2, 3]
y3 = [1, 2, 3]
x3.append(4)

print("y3: ", y3)
print("x3: ", x3)

print(x3 is y3)
```

# Membership Operators

- these are used to test whether a value is present in a sequence(strings, list, tuple, set and dictionary)

```python
msg = "Hello World"
print('H' in msg)
print('z' in msg)
print('m' not in msg)
```

```python
l = [23, 79, 56, 42, 37]

if 33 in l:
    print("element is presenyt in list")
else:
    print("not present")
```
